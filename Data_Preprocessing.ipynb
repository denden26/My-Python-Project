Pada Project kali ini sebagai Data Scientist di ID/X Partners saya mendapatkan tugas yaitu membersihkan data atau data preprocessing. Data preprocessing ini merupakan tahapan setelah data collection yang nantinya akan digunakan pada permodelan data.

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing

app_record = pd.read_csv('/content/application_record.csv')
app_record.head()
```

**Melakukan one hot encoding pada kolom CODE_GENDER untuk membuat dummy**

```python
app_record[['CODE_GENDER']].head()
one_hot=pd.get_dummies(app_record['CODE_GENDER'], drop_first=True)
one_hot.head()
```

**Melakukan label encoding pada kolom NAME_EDUCATION_TYPE**

```python
label_encoder = preprocessing.LabelEncoder()
app_record['NAME_EDUCATION_TYPE_enc']= label_encoder.fit_transform(app_record['NAME_EDUCATION_TYPE'])
app_record[['NAME_EDUCATION_TYPE','NAME_EDUCATION_TYPE_enc']].head()
```

**Mengubah nilai pada kolom FLAG_PHONE: 0 menjadi N dan 1 menjadi Y**

```python
app_record[['FLAG_PHONE']].head()
```

```python
app_record['FLAG_PHONE']= np.where(app_record['FLAG_PHONE'] == 1,'Y','N')
app_record[['FLAG_PHONE']].head()
```

```python
app_record[['FLAG_PHONE']].head()
app_record['FLAG_PHONE']= np.where(app_record['FLAG_PHONE'] == 0,'N','Y')
app_record[['FLAG_PHONE']].head()
```

```python
app_record.head(10)
```

```python
app_record[['AMT_INCOME_TOTAL']].head()
```

***Menstandarisasi kolom AMT_INCOME_TOTAL***

```python
ss = preprocessing.StandardScaler()

app_record['AMT_INCOME_TOTAL_scaled'] = ss.fit_transform(app_record[['AMT_INCOME_TOTAL']])

app_record[['AMT_INCOME_TOTAL','AMT_INCOME_TOTAL_scaled']].head()
```

**Membulatkan kebawah angka pada AGE**

```python
app_record[['DAYS_BIRTH']].head()
```

```python
app_record['AGE'] = app_record['DAYS_BIRTH'] / -365
app_record['AGE'] = app_record['AGE'].apply(np.floor)
app_record[['DAYS_BIRTH','AGE']].head()
```

**Mengganti nilai kosong (NULL) pada kolom OCCUPATION_TYPE menjadi "Unknown"**

```python
app_record['OCCUPATION_TYPE'].isnull().sum()
```

```python
app_record['OCCUPATION_TYPE'] = app_record['OCCUPATION_TYPE'].fillna('Unknown')
app_record['OCCUPATION_TYPE'].isnull().sum()
```

```python
app_record[['OCCUPATION_TYPE']].head()
```

```python
app_record['OCCUPATION_TYPE'].fillna('Unknown', inplace=True)
```

```python
app_record[['OCCUPATION_TYPE']].head()
```

**Mengubah tipe data kolom ID dari integer menjadi object (str)**

```python
type(app_record.ID.values[0])
```

```python
app_record.ID = app_record.ID.astype(str)
type(app_record.ID.values[0])
```

**Membuang kolom ID dan CODE_GENDER dari dataframe app_record**

```python
app_record.drop(['ID', 'CODE_GENDER'], axis=1, inplace=False)
```
